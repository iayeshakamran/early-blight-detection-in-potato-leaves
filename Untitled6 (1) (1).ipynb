```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, LeakyReLU, BatchNormalization, Reshape, Conv2DTranspose, Conv2D, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# Define image dimensions
image_width = 32
image_height = 32

# Define batch size
batch_size = 32

# Load your dataset using ImageDataGenerator
train_data_gen = ImageDataGenerator(rescale=1./255)

# Training data generator
train_path = 'pimages'  # Update the path to the 'pimages' directory
train_generator = train_data_gen.flow_from_directory(
    train_path,
    target_size=(image_width, image_height),
    batch_size=batch_size,
    class_mode='categorical',  # Use categorical class mode for multi-class classification
    shuffle=True
)

# Get the number of classes from the generator
num_classes = len(train_generator.class_indices)

# Define the GAN model architecture for zero-shot learning
def create_gan_model(input_shape, num_classes):
    # Generator model
    latent_dim = 100
    generator_input = Input(shape=(latent_dim,))
    x = Dense(128 * 8 * 8)(generator_input)
    x = LeakyReLU()(x)
    x = Reshape((8, 8, 128))(x)
    x = Conv2DTranspose(128, kernel_size=4, strides=2, padding='same')(x)
    x = LeakyReLU()(x)
    x = Conv2DTranspose(256, kernel_size=4, strides=2, padding='same')(x)
    x = LeakyReLU()(x)
    x = Conv2D(3, kernel_size=3, activation='sigmoid', padding='same')(x)  # Output image with 3 channels
    generator = Model(generator_input, x)

    # Discriminator model
    discriminator_input = Input(shape=input_shape)
    x = Conv2D(64, kernel_size=3, strides=2, padding='same')(discriminator_input)
    x = LeakyReLU()(x)
    x = Conv2D(128, kernel_size=3, strides=2, padding='same')(x)
    x = LeakyReLU()(x)
    x = Flatten()(x)
    x = Dense(num_classes, activation='softmax')(x)  # Output classification probabilities for each class
    discriminator = Model(discriminator_input, x)

    # Compile the discriminator
    discriminator.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.0002, beta_1=0.5), metrics=['accuracy'])

    # Combine generator and discriminator to create GAN model
    gan_input = Input(shape=(latent_dim,))
    gan_output = discriminator(generator(gan_input))
    gan_model = Model(gan_input, gan_output)
    gan_model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.0002, beta_1=0.5))

    return gan_model, generator, discriminator

# Create the GAN model
latent_dim = 100  # Size of the latent space
input_shape = (image_width, image_height, 3)  # Change to (image_width, image_height, num_channels) if needed
gan_model, generator_model, discriminator_model = create_gan_model(input_shape, num_classes)

# Training loop for the GAN
epochs = 100
discriminator_losses = []
generator_losses = []
discriminator_accuracies = []
for epoch in range(epochs):
    print(f"Epoch {epoch + 1}/{epochs}")

    # Training the discriminator
    for _ in range(1):  # You can adjust the number of iterations here
        # Get a batch of real images and generate fake images
        real_images, labels = train_generator.next()  # Use the train_generator to get real images and labels
        latent_space = np.random.normal(size=(batch_size, latent_dim))  # Generate random noise vectors
        generated_images = generator_model.predict(latent_space)  # Get the generator from the GAN model

        # Combine real and generated images, and their labels
        images = np.concatenate([real_images, generated_images])
        labels = np.concatenate([labels, np.zeros((batch_size, num_classes))])  # Label the fake images as zeros

        # Train the discriminator with real and generated images
        discriminator_loss, discriminator_accuracy = discriminator_model.train_on_batch(images, labels)
        discriminator_losses.append(discriminator_loss)
        discriminator_accuracies.append(discriminator_accuracy)

    # Training the generator
    latent_space = np.random.normal(size=(batch_size, latent_dim))  # Generate random noise vectors
    valid_labels = np.ones((batch_size, num_classes))  # Label the generated images as ones

    # Train the generator to fool the discriminator
    generator_loss = gan_model.train_on_batch(latent_space, valid_labels)
    generator_losses.append(generator_loss)

    print(f"Discriminator Loss: {discriminator_loss}, Discriminator Accuracy: {discriminator_accuracy} | Generator Loss: {generator_loss}")

# After the training loop, you can use the generator part of the GAN to generate images
# using random noise vectors for zero-shot learning.

# Plot the GAN training losses and discriminator accuracy
plt.figure(figsize=(10, 5))
plt.plot(range(1, epochs + 1), discriminator_losses, label='Discriminator Loss')
plt.plot(range(1, epochs + 1), generator_losses, label='Generator Loss')
plt.plot(range(1, epochs + 1), discriminator_accuracies, label='Discriminator Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Loss / Accuracy')
plt.title('GAN Training Losses and Discriminator Accuracy')
plt.legend()
plt.show()
```
